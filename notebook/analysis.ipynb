```python
## Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("customer_segmentation.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.describe()
```

```python
df.columns
```

```python
df['Marital_Status'].value_counts()
```

```python
df['Education'].value_counts()
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'], dayfirst=True)
```

```python
df.info()
```

```python
df['Age'] = 2025 - df['Year_Birth']
df['Age'].head()
```

```python
df.head()
```

```python
df['Total_Children'] = df['Kidhome'] + df['Teenhome']
df['Total_Children'].head()
```

```python
df.columns
```

```python
spend_columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
```

```python
df['Total_Spend'] = df[spend_columns].sum(axis=1)
df['Total_Spend'].head()
```

```python
df.head()
```

```python
df['Customer_Since_Days'] = (pd.to_datetime('today') - df['Dt_Customer']).dt.days
```

```python
df['Customer_Since_Days'].sample(5)
```

```python
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.histplot(df['Income'], bins=30, kde=True)
plt.title('Income Distribution')
plt.show()
```

```python
sns.histplot(df['Total_Spend'], bins=30, kde=True)
plt.title('Total Spend Distribution')
plt.show()
```

```python
sns.boxplot(x='Education', y='Income', data=df)
plt.xticks(rotation=45)
plt.title('Income by Education Level')
plt.show()
```

```python
sns.boxplot(x='Marital_Status', y='Total_Spend', data=df)
plt.title('Total Spend by Marital Status')
plt.show()
```

```python
df.columns
```

```python
corr = df[['Income', 'Total_Spend', 'Age', 'Recency', 'NumWebPurchases', 'NumStorePurchases']].corr()
```

```python
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
pivot_income = df.pivot_table(values='Income', index='Education', columns='Marital_Status', aggfunc='mean')
```

```python
pivot_income
```

```python
sns.heatmap(pivot_income, annot=True, fmt=".0f", cmap='YlGnBu')
plt.title('Average Income by Education and Marital Status')
plt.show()
```

```python
group1 = df.groupby('Education')['Total_Spend'].mean().sort_values(ascending=False)
group1
```

```python
group1.plot(kind='bar', color='skyblue')
plt.title('Average Total Spend by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Average Total Spend')
plt.xticks(rotation=45)
plt.show()
```

```python
df.columns
```

```python
df['AcceptedAny'] = df[['AcceptedCmp1', 'AcceptedCmp2', 'AcceptedCmp3', 'AcceptedCmp4', 'AcceptedCmp5', 'Response']].sum(axis=1)
```

```python
df['AcceptedAny'].unique()
```

```python
df['AcceptedAny'] = df['AcceptedAny'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df['AcceptedAny'].unique()
```

```python
group2 = df.groupby("Marital_Status")['AcceptedAny'].mean().sort_values(ascending=False)
```

```python
group2
```

```python
group2.plot(kind='bar', color='coral')
plt.title('Proportion of Customers Accepting Any Campaign by Marital Status')
plt.xlabel('Marital Status')
plt.ylabel('Proportion Accepting Any Campaign')
plt.show()
```

```python
bins = [18, 30, 40, 50, 60, 70, 90]
labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']
```

```python
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
```

```python
df['AgeGroup'].head()
```

```python
group3 = df.groupby('AgeGroup')['Income'].mean()
```

```python
group3
```

```python
group3.plot(kind='barh', color='lightgreen')
plt.title('Average Income by Age Group')
plt.xlabel('Average Income')
plt.ylabel('Age Group')
plt.show()
```
